```python
# Dependencies and Setup
import pandas as pd
import numpy as np
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(50)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_number_schools_list=school_data_complete["school_name"].unique()

total_number_schools=school_data_complete["school_name"].nunique()

total_number_students=school_data_complete["Student ID"].count()
total_number_students

budget=school_data_complete["budget"].unique()
total_budget=budget.sum()
total_budget

avg_mth=school_data_complete["math_score"].mean()
avg_mth

avg_reading=school_data_complete["reading_score"].mean()
avg_reading

passing_math=school_data_complete.loc[school_data_complete["math_score"]>=70]
passing_math_count=passing_math.count()
percent_passing_math=(passing_math_count/total_number_students)*100
percent_passing_math

passing_reading=school_data_complete.loc[school_data_complete["reading_score"]>=70]
passing_reading_count=passing_reading.count()
percent_passing_reading=(passing_reading_count/total_number_students)*100
percent_passing_reading

overall_passing_rate=(avg_reading+avg_mth)/2
overall_passing_rate
district_summary_df=pd.DataFrame({
    "Total Schools":[total_number_schools],
    "Total Students":[total_number_students],
    "Total Budget($)":[total_budget],
    "Average Math Score":[avg_mth],
    "Average Reading Score":[avg_reading],
    "Percent Passing Math (%)":[percent_passing_math],
    "Percent Passing Reading (%)":[percent_passing_reading],
    "Percent Overall Passing Rate (%)":[overall_passing_rate]
})
district_summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_data_complete["passing_math"]=school_data_complete["math_score"]>=70
school_data_complete["passing_reading"]=school_data_complete["reading_score"]>=70
new_school_data=school_data_complete.groupby(["school_name"]).mean()
new_school_data.head()

new_school_data = pd.merge(new_school_data, school_data, how="left", on=["school_name", "school_name"])


new_school_data.head()

del new_school_data["School ID_x"]
del new_school_data["School ID_y"]
del new_school_data["Student ID"]
del new_school_data["budget_y"]
del new_school_data["size_y"]

new_school_data.head()

percent_budget=new_school_data["budget_x"]/new_school_data["size_x"]
percent_budget.head()

new_school_data["Per Student Budget"]=percent_budget

new_school_data.head()

new_school_data["Percent Passing Math(%)"]=new_school_data["passing_math"]*100

new_school_data.head()
new_school_data["Percent Passing Reading(%)"]=new_school_data["passing_reading"]*100
new_school_data.head()

del new_school_data["passing_math"]
new_school_data.head()

del new_school_data["passing_reading"]
new_school_data.head()

new_school_data["Percent Overall Passing Rate (%)"]=(new_school_data["Percent Passing Reading(%)"]+new_school_data["Percent Passing Math(%)"])/2
new_school_data.head()

new_school_data_df=pd.DataFrame({"School Name":  new_school_data["school_name"],
                                "School Type": new_school_data["type"],
                                "Total Students":new_school_data["size_x"],
                                "Total School Budget": new_school_data["budget_x"],
                                "Per Student Budget":new_school_data["Per Student Budget"],
                                "Average Math Score":new_school_data["math_score"],
                                "Average Reading Score":new_school_data["reading_score"],
                                "Percent Passing Math(%)":new_school_data["Percent Passing Math(%)"],
                                "Percent Passing Reading (%)":new_school_data["Percent Passing Reading(%)"],
                                "Percent Overall Passing Rate (%)":new_school_data["Percent Overall Passing Rate (%)"]
                                  })
new_school_data_df.head()


revised_nsdf=new_school_data_df.set_index("School Name")
revised_nsdf.head()

top_performing=revised_nsdf.sort_values(["Percent Overall Passing Rate (%)"],ascending=False)
top_performing.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_performing=revised_nsdf.sort_values("Percent Overall Passing Rate (%)",ascending=True)
bottom_performing.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth=  school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth =  school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh =  school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelveth=  school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["math_score"]

per_grade_math_df=pd.DataFrame({"9th":ninth,
                               "10th":tenth,
                               "11th":eleventh,
                               "12th":twelveth})

per_grade_math_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
spending_df=revised_nsdf
spending_df.head()

spending_df["School Spending"]=pd.cut(spending_df["Per Student Budget"],spending_bins,labels=group_names)

spending_df.head()

spending_df=spending_df.set_index("School Spending")
spending_df.head()

del spending_df["School Type"]
del spending_df["Total Students"]
del spending_df["Per Student Budget"]
del spending_df["Total School Budget"]
spending_df.head()
```
